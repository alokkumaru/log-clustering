```python
import pandas as pd
import gzip
import os
from elasticsearch import Elasticsearch
from pandasticsearch import Select
from update import update_info, update_err
import clustering
```

### Opening log files

```python
#reading lines from path that contains all zipped log files
lines=[]
for file in os.listdir():
    #print(file)
    if file.endswith(".gz"):
        with gzip.open(file, 'rb') as f:
            lines=lines+f.read().splitlines()
            f.close()
```

### Reading log files into dataframes

```python
df1=pd.DataFrame({'logs':lines})
df1=df1['logs'].str.decode("utf-8")
df1=df1.str.split(expand=True)
```

```python
cols=[i for i in range(len(df1.columns))]
cols=cols[6:]
df1[cols]=df1[cols].replace([None],'')
df1[5] = df1[5].str.cat(df1[cols].astype(str), sep=" ")
df1 = df1.drop(cols, 1)
df1.columns=['date','time','severity','pid','process','message']
```

```python
error_df=df1[df1['severity']=='[ERROR]']
info_df=df1[df1['severity']=='[INFO]']
#print(error_df.query("pid=='6759'")['logmsg'][2282])
#len(error_df)
```

```python
log_info=clustering.unsupervised_clustering_info(info_df,number_of_clusters=16)
log_err=clustering.unsupervised_clustering_err(error_df,number_of_clusters=2)
log=pd.concat([log_info,log_err])
```
